```python
from forex_python.bitcoin import BtcConverter
b = BtcConverter ()
```

```python
from datetime import datetime

start = datetime(2015,1,1)
end = datetime(2021,12,27)
```

```python
price_data = b.get_previous_price_list("USD", start, end)
```

```python
len(price_data)
```

```python
type(price_data)
```

```python
#Reading into Pands dataframe
```

```python
import pandas as pd

prices = pd.DataFrame(price_data.items(), columns=["date", "price"])
```

```python
prices.head()
```

```python
#convert date column to index

prices = prices.set_index('date')
prices.head()
```

```python
prices.loc["2017-01-01":"2017-12-31"]
```

VISUALIZATION AND PLOTTING

```python
!pip install matplotlib
prices.plot()
```

```python
prices.rolling(30).mean().plot()
```

```python
import matplotlib.style as style

style.use('fivethirtyeight')
rolling = prices.rolling(30).mean()
rolling.plot()
```
