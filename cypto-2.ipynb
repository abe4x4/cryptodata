<h4>GETTING THE DATA</h4>
Download hourly interval data for the whole 2021 year for 5 cryptocurrencies: Bitcoin (BTC), Ethereum (ETH), Litecoin (LTC), Maker (MKR) and Binance (BNB) from Yahoo Finance API (yfinance)

```python
import yfinance as yf
import pandas as pd
import seaborn as sns

# Get BITCOIN Historical Pricing Data in 2021 by hour
btc = yf.Ticker('BTC-USD')
btc_data = btc.history(interval='1h', start='2021-01-01', end='2021-12-31')

# Get ETHEREUM 2021 data by hour
eth = yf.Ticker('ETH-USD')
eth_data = eth.history(interval='1h', start='2021-01-01', end='2021-12-31')

# Get DASH 2021 data by hour
dash = yf.Ticker('DASH-USD')
dash_data = dash.history(interval='1h', start='2021-01-01', end='2021-12-31')

# Get Litecoin Inu Historical Pricing Data in 2021 by hour
ltc = yf.Ticker('LTC-USD')
ltc_data = ltc.history(interval='1h', start='2021-01-01', end='2021-12-31')

# Get BINANCE Historical Pricing Data in 2021 by hour
bnb = yf.Ticker('bnb-USD')
bnb_data = bnb.history(interval='1h', start='2021-01-01', end='2021-12-31')
```

<h4>DATA MANIPULATION</h4>

```python
#Extract hourly closing prices for 5 coins (BTC, ETH, SOL, DOT, and BNB), then merge data into pandas dataframe (df)
import pandas as pd
df = pd.DataFrame({'BTC': btc_data.Close,
                   'ETH': eth_data.Close,
                   'DASH': dash_data.Close,
                   'LTC': ltc_data.Close,
                   'BNB': bnb_data.Close})
```

<h4>BUY AND HOLD STRATEGY</h4>
Buy and hold is a passive investment strategy in which an investor buys a cryptocurrency and holds it for a relatively long period, regardless of fluctuations in the market.

Letâ€™s analyze returns using the Buy and hold strategy in 2021. We calculate the return percentage, where t represents a certain period and price0 is the initial closing price:



```python
df_return = df.apply(lambda x: x / x[0])
df_return.head()
```

<h4>Returns Visualization</h4>
It shows that BNB was the most profitable for the period between January 1st, 2021 and December 31st, 2021

```python
df_return.plot(grid=True, figsize=(15, 10)).axhline(y = 1, color = "black", lw = 2)
```

What is the percentage increase?

```python
df_perc = df_return.tail(1) * 100
ax = sns.barplot(data=df_perc)
df_perc
```

How many coins could we bought for $1000?

```python
budget = 1000 # USD
df_coins = budget/df.head(1)

ax = sns.barplot(data=df_coins)
df_coins
```

How much money would we make?

```python
df_profit = df_return.tail(1) * budget

ax = sns.barplot(data=df_profit)
df_profit
```
